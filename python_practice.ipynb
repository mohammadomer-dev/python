### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
#take user's salary
salary=int(input("Enter Your CTC(IN Lakhs):"))
salary=100000*salary
print("Salary after All Deduction")
if salary<500000:
  print("No Tax Deduction")
elif salary>=500000 and salary<1000000:
  salary=salary-salary*0.1
  print("Tax 10%=",salary)
elif salary>=1000000 and salary<2000000:
  salary=salary-salary*0.2
  print("Tax 20%=",salary)
else:
  salary=salary-salary*0.3
  print("Tax 30%=",salary)


salary=salary-salary*0.1
print("HRA 10%=",salary)
salary=salary-salary*0.05
print("DA 5%=",salary)
salary=salary-salary*0.03
print("PF 3%=",salary)

salary=salary/12
print("In Hand Monthly Salary=",salary)

````

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Write code here
no = int(input("Enter number: "))
rev = 0

while no > 0:
    digit = no % 10
    rev = rev * 10 + digit
    no = no // 10

print("Reverse =", rev)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
n = int(input("Enter N: "))

i = 1
total = 0

while i <= n:

    if i % 5 == 0:
        i += 1
        continue

    if total + i > 300:
        break

    total += i
    i += 1

print(total)
````

###`Problem 12`:Write a program to print whether a given number is a prime number or not

````python
# Write code here
a = int(input("Enter any number: "))

prime = True

for i in range(2, a):
    if a % i == 0:
        prime = False
        break

if prime:
    print("It's a Prime no.")
else:
    print("It's not a Prime no.")
````
